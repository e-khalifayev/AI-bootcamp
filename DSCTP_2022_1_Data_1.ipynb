#### 1. Read the two cars datasets

```python
import pandas as pd

df1 = pd.read_csv('cars1.csv')
df2 = pd.read_csv('cars2.csv')

print(df1)
print(df2)
```

#### 2. Remove the blank columns from the first dataset

```python
df1.dropna(how = 'all', axis = 1, inplace = True)

print(df1)
```

#### 3. Join two datasets

```python
df = pd.concat([df1, df2])
print(df)
```

#### 4. Find the number of cars weighing more than 3000

```python
count = df['weight'][df['weight'] > 3000].count()
print(count)
```

#### 5. Get the rows with "Ford" in its name

```python
ford = df[df['car'].str.contains('ford')]
```

```python
print(ford)
```

#### 6. Find the maximum acceleration for each number of cylinders

```python
df.groupby(['cylinders']).agg(max_acceleration = ('acceleration', 'max'))
```

```python
df.groupby(['cylinders'])['acceleration']
```

#### 8. Find the minimum, 25th percentile, median, 75th, and maximum for the weight

```python
print("min: ", df['weight'].min())
print("25th percentile: ", df['weight'].quantile(0.25))
print("median: ", df['weight'].quantile(0.50))
print("75th percentile: ", df['weight'].quantile(0.75))
print("max: ", df['weight'].max())

print()
#or...
df['weight'].describe()
```

#### 9. Bin the weight into quartiles and replace the values with the bin name

```python
bindf = pd.cut(x = df['weight'],\
               bins = [df['weight'].min(), df['weight'].quantile(0.25),\
                df['weight'].quantile(0.50), df['weight'].quantile(0.75),\
                df['weight'].max()],\
               labels = ['q1', 'q2', 'q3', 'q4'])

```

```python
print(bindf)
```
